1. Implementing Linear Regression


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Sample Data
# Generating random data for demonstration
np.random.seed(0)
X = np.random.rand(100, 1) * 10
y = 2.5 * X + np.random.randn(100, 1) * 2

# Split the data into training/testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Create linear regression object
regr = LinearRegression()

# Train the model using the training sets
regr.fit(X_train, y_train)

# Make predictions using the testing set
y_pred = regr.predict(X_test)

# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean squared error
print('Mean squared error: %.2f' % mean_squared_error(y_test, y_pred))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f' % r2_score(y_test, y_pred))
```

2. Implementing Decision Tree with Hyperparameter Tuning


```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

# Sample Data
# Generating random data for demonstration
X = np.random.rand(100, 5)
y = np.random.randint(0, 2, size=100)

# Split the data into training/testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Create the parameter grid based on the results of random search
param_grid = {
    'max_depth': [3, 5, 10, None],
    'max_features': [1, 3, 10],
    'min_samples_split': [2, 3, 10],
    'min_samples_leaf': [1, 3, 10],
    'criterion': ['gini', 'entropy']
}

# Create a base model
dt = DecisionTreeClassifier()

# Instantiate the grid search model
grid_search = GridSearchCV(estimator=dt, param_grid=param_grid,
                          cv=3, n_jobs=-1, verbose=2)

# Fit the grid search to the data
grid_search.fit(X_train, y_train)

print("Best parameters found: ", grid_search.best_params_)
best_grid = grid_search.best_estimator_
y_pred = best_grid.predict(X_test)
```

3. Implementing Naive Bayes


```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# Sample Data
# Generating random data for demonstration
X = np.random.rand(100, 5)
y = np.random.randint(0, 2, size=100)

# Split the data into training/testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Create a Gaussian Classifier
model = GaussianNB()

# Train the model using the training sets
model.fit(X_train, y_train)

# Predict the response for test dataset
y_pred = model.predict(X_test)

# Model Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))
```

4. Implementing Support Vector Machine and Finding Metrics


```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix

# Sample Data
# Generating random data for demonstration
X = np.random.rand(100, 5)
y = np.random.randint(0, 2, size=100)

# Split the data into training/testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Create a SVM Classifier
model = SVC(kernel='linear') # Linear Kernel

# Train the model using the training sets
model.fit(X_train, y_train)

# Predict the response for test dataset
y_pred = model.predict(X_test)

# Model Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))

# Confusion Matrix
print(confusion_matrix(y_test, y_pred))

# Classification Report
print(classification_report(y_test, y_pred))
```

5. Differences between Descriptive Statistics and Inferential Statistics

1.Purpose: Descriptive statistics summarize and describe the features of a dataset, while inferential statistics use sample data to make inferences about a population.

2.Data Description: Descriptive statistics focus on describing data through measures like mean, median, and mode, whereas inferential statistics use data to estimate population parameters and test hypotheses.

3. Usage: Descriptive statistics are used to present the raw data in a meaningful way, while inferential statistics are used to draw conclusions beyond the immediate data.

4. Population vs. Sample: Descriptive statistics describe the entire dataset or a sample, while inferential statistics aim to make predictions or inferences about a population based on a sample.

5. Techniques: Techniques include mean, median, mode, standard deviation for descriptive statistics; hypothesis testing, confidence intervals, and regression analysis for inferential statistics.

6. Outcome: Descriptive statistics provide straightforward, summary results, while inferential statistics provide probabilistic statements about population parameters.

7.Assumptions: Inferential statistics rely on assumptions about the data distribution and sampling, whereas descriptive statistics do not.

8 Complexity: Descriptive statistics are generally simpler and easier to compute, whereas inferential statistics involve more complex calculations.

9 Role: Descriptive statistics are the first step in data analysis, providing initial insights, whereas inferential statistics are used for deeper analysis and decision-making.

10. Examples: Examples of descriptive statistics are calculating the mean or creating a histogram; examples of inferential statistics are t-tests, chi-square tests, and ANOVA.




6. Handling Missing Values in Machine Learning

```python
import pandas as pd
from sklearn.impute import SimpleImputer

# Sample Data
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [np.nan, 2, 3, 4, 5],
    'C': [1, 2, 3, 4, np.nan]
}
df = pd.DataFrame(data)

# Using SimpleImputer to fill missing values with the mean
imputer = SimpleImputer(strategy='mean')
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)

print(df_imputed)
```

7. Plotting Visualizations


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sample Data
data = sns.load_dataset("iris")

# 1. Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x="sepal_length", y="sepal_width", hue="species")
plt.title('Scatter plot of Sepal Length vs Sepal Width')
plt.show()

# 2. Histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x="petal_length", kde=True)
plt.title('Histogram of Petal Length')
plt.show()

# 3. Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x="species", y="petal_width")
plt.title('Box plot of Petal Width by Species')
plt.show()
```
